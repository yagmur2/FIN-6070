### h. Functions
A function is a block of reusable code that is used to perform a job and allows for it to be repeated at other points within your program. For example, you can write a function to calculate the dividend yield or bid/ask spread. Functions help to break your code into smaller parts. Python provides you many built-in functions such as print(), and you can also create your own functions which are called <b>user-defined</b> functions.

To define functions in Python, follow the principles given below:
- Function blocks start with the def keyword followed by the function name and round brackets '( )'
- Input parameters must be placed in these round brackets
- The block of code in the function is indented and is preceded by a colon (:)


    def function_name(input_parameters):

        statement(s)

        return [expression or variable]

````python
# lets start by defining a new function to calculate bid/ask spreads

def ba_spread(bid,ask): # function takes 2 input values, a bid and an ask
    spread = ask - bid # we calculate the spread as ask minus bid
    return spread # we return the value of spread
````

````python
# we can also calculate the mid price of the bid/ask spread

def mid_price(bid,ask):
    mid = (bid + ask) / 2
    return mid
````

````python
# it is possible we just want to print the values

def print_values(bid,ask):
    spread = ba_spread(bid,ask)
    mid = mid_price(bid,ask)
    for i in [spread, mid]: # use a loop to print each value individually
        print(i)
````

````python
# we can test our functions with some made up values.

b = 5
a = 5.1
````

````python
# Computing transaction charges and PnL
# Assumptions: Transactions charges of 2% of buying price, 15% tax on profits.

def absolute_PnL(buy_price, sell_price):
    PnL = sell_price - buy_price
    return PnL
````

````python
def transaction_charges(buy_price):
    trans_charges = buy_price * 0.02
    return trans_charges
````

````python
def taxes(buy_price, sell_price):
    profit = sell_price - buy_price
    if profit > 0:
        tax = 0.15 * profit
    else:
        tax = 0
    return tax
````

````python
bought = 100
sold = 125
PnL = absolute_PnL(bought, sold) - transaction_charges(bought) - taxes(bought, sold)
print('We made $' + str(PnL), 'after considering transaction costs and our tax bill.')
````

## 5. Data preprocessing
We will cover some of the useful data preprocessing methods like data cleaning, outlier detection, and data resampling.

### a. Data cleaning
When you are working with raw data, instances of duplicate data, missing data, or inconsistent data can occur. If the data is not cleaned then your analysis can give incorrect results.

````python
# Duplicate data means the same data values getting repeated for certain timestamps. 
# We can detect the occurrence of duplicate timestamps using the duplicated method 
# and remove them using the drop_duplicates method.

df = pd.read_csv('duplicate_data.csv', index_col='DATE', 
                 usecols=['DATE', 'OPEN', 'HIGH', 'LOW', 'CLOSE'], parse_dates=True)
df
````

````python
# Use the duplicated method to highlight the repeated timestamps

df1 = df[df.duplicated(keep = False)] # False displays all of the duplicate value, first will show only the first
df1
````

````python
# Use the drop_duplicates method to remove the duplicate timestamps.
# In this example, it will remove the duplicate values for 09:16:00 and 09:20:00

df.drop_duplicates(inplace = True)
df
````

````python
# Missing data is another data quality issue. 
# On time frames like minute or daily data we can build a check in the code to find missing data.

df = pd.read_csv('missing_data.csv', index_col='DATE',
                 usecols=['DATE', 'OPEN', 'HIGH', 'LOW', 'CLOSE'], parse_dates=True)
df
````

````python
# In Python, we can use the reindex method to detect missing data for the specific frequency. 
# The NaN values can be dropped or filled using the previous values or by any other method.

df.reindex(pd.date_range(start=df.index[0], end=df.index[-1], freq='1min'))
````

````python
# Inconsistent data can occur in different forms, for example, a spike (outlier) in the price series or volume. 
# Checks in the code can help detect outliers.

df = pd.read_csv('outlier_data.csv', index_col='DATE',
                 usecols=['DATE', 'OPEN', 'HIGH', 'LOW', 'CLOSE'], parse_dates=True)
df
````

In the above table that there is a spike in the close price at timestamp of 09:18:00. To detect such spikes we can use a simple diff() method on the close price of the dataframe and specify the threshold. This method assumes domain knowledge where you can choose a value as the cutoff. Other methods include statistically determining outliers based on their distance from a median or mean value. In the example below, we have specified threshold value of 25 to classify it as a spike. 

````python
df['CLOSE'].diff() > 25
````

````python
# The Pandas ‘resample’ function can be used to convert to desired time frames like minutely, hourly, daily, or weekly. 
# We will illustrate how to convert a 1-minute time series into a 5-minute time series.

df1 = pd.read_csv('1_min.csv', index_col='DATE',
                  usecols=['DATE', 'OPEN', 'HIGH', 'LOW', 'CLOSE'], parse_dates=True)
df1.head(10)
````

We use the resample function which takes two arguments to create a resampled series. The first argument, &#39;5Min&#39; specifies the time period to resample and the second argument, &#39;label&#39; determines which bin edge label to label bucket with.

Since we are dealing with OHLC (Open, High, Low, Close) data, we would want to have the same format and meaning of OHLC in the resampled series. This means that the &#39;OPEN&#39; price in the resampled series needs to correspond to the open price of the first 1-minute bar out of the five 1-minute bars used for resampling. The &#39;HIGH&#39; in the resampled series needs to be equal to the highest price of the five 1-minute bars. Similarly, the &#39;LOW&#39; in the resampled series needs to be equal to the lowest price of the five 1-minute bars. The &#39;CLOSE&#39; price in the resamples series needs to correspond to the close price of the last 1-minute bar of the five 1-minute bars used for resampling.

This is achieved by using the aggregate method and specifying the requirements explained above to the aggregate method in the form of a Python dictionary.

````python
from IPython.display import display_html

df2 = df1.resample('5Min',label='right').agg({'OPEN': 'first','HIGH':'max','LOW':'min','CLOSE':'last'})
df2.head(8)

df1_styler = df1.style.set_table_attributes("style='display:inline'").set_caption('Table 1 - 1 Minute Data')
df2_styler = df2.style.set_table_attributes("style='display:inline'").set_caption('Table 2 - 5 Minute Aggegregated Data')

display_html(df1_styler._repr_html_()+df2_styler._repr_html_(), raw=True)

## Table 1 shows the original 1-minute price series and the table 2 on the right shows the 5-minute resamples series.
````

## 7. Extra if inerested: Performance metrics of Buy and Hold strategy
We will analyse the performance of a trading strategy using various performance metrics.

````python
# download prices
stock_data =  yf.download(tickers = 'AAPL AMZN WMT KO SNAP', start = '2016-01-01', end = '2020-12-31')['Adj Close']
stock_data.head(40)
````

````python
# convert daily data to monthly data
monthly_stock_data = stock_data.asfreq('M')
#monthly_stock_data = stock_data.asfreq('M', method='ffill')
monthly_stock_data.head()
````

````python
# calculate returns
monthly_stock_data = monthly_stock_data.pct_change()
monthly_stock_data = monthly_stock_data.dropna()
monthly_stock_data.head()
````

````python
# calculate portfolio returns
monthly_port_ret = monthly_stock_data.mean(axis = 1) # average portfolio return (ie. left to right across columns)

# calculate cumulative performance

port_ret = (monthly_port_ret + 1).cumprod()-1
port_ret.head()
````

````python
# Import matplotlib as an alias plt and set the style
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-darkgrid')

# Plot cumulative strategy returns
port_ret.plot(figsize=(10, 7))

# Set title and labels for the plot
plt.title('Buy and Hold Strategy Returns', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Cumulative Returns', fontsize=12)
plt.show()
````

````python
# Total Returns

total_returns = (port_ret.iloc[-1])*100

print('The total returns of strategy is %.2f' % total_returns + '%')
````

````python
# Annualized Returns

# Total number of trading days
months = len(port_ret)

# Calculate the compounded annual growth rate
cagr = (port_ret.iloc[-1]**(12/months)-1)*100
print('The CAR of strategy is %.2f' % cagr + '%')
````

````python
# Sharpe Ratio (Return - RF) / Standard Deviation

# Assume that an average annual risk-free rate is 1%
risk_free_rate = 0.01/12

# Calculate and annualize the Sharpe Ratio
sharpe = (np.mean(monthly_port_ret) - (risk_free_rate))/np.std(monthly_port_ret)
print("The Sharpe ratio of strategy is %.2f" % sharpe)
````
